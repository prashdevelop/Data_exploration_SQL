### 3.6 Clean & Add Columns for Analysis


#### 3.6.1 Date formatting: Using 'format()'

The 'date' columns on both dataframes (as can be seen at the 'head()' step - section 3.4.1) are wrongly classified as "Characters". 'format()' is used to change the formats to datetime:

```r
daily_activity$ActivityDate=as.POSIXct(daily_activity$ActivityDate, format="%m/%d/%Y", tz=Sys.timezone())
daily_activity$date <- format(daily_activity$ActivityDate, format = "%m/%d/%y")

sleep_day$SleepDay=as.POSIXct(sleep_day$SleepDay, format="%m/%d/%Y %I:%M:%S %p", tz=Sys.timezone())
sleep_day$date <- format(sleep_day$SleepDay, format = "%m/%d/%y")
```

#### 3.6.2 Adding New Columns: Using 'mutate ()'

To aid analysis and visualization, columns are added to both dataframes using the mutate() function:

* 'daily_activity' dataframe

For the dataframe 'daily_activity', there is no column showing the total active minutes,

* 'sleep_day' dataframe

For the dataframe 'sleep_day', there is no column showing total active minutes while in bed

```r
daily_activity <- mutate(daily_activity, TotalActiveMinutes = VeryActiveMinutes + FairlyActiveMinutes + LightlyActiveMinutes)

sleep_day <- mutate(sleep_day, TotalMinutesAwake = TotalTimeInBed - TotalMinutesAsleep)
```

#### 3.6.3 Summary statistics: Using 'summary()'

The summary() function returns a wide range of summary statistics.

```r
daily_activity %>%  
  select(TotalSteps,
         TotalDistance,
         SedentaryMinutes, 
         TotalActiveMinutes,
         Calories) %>%
  summary()
```

The following can be inferred from this summary (especially using the mean/average values as yardstick):

* The Academy of Nutrition and Dietetics recommends slow weight loss for lasting results—usually 1/2 pound to one pound per week. One pound of body fat equals 3500 calories, so depending on your weight and workout intensity, you could lose about one pound per week simply by completing an extra 10,000 steps each day. The mean total steps taken is 7638 which means more than half of the sample population do not do enough activity to lose up to 1LB per week.
* Also, 3500 extra calories each week comes to 500 extra calories daily. On average, American adults eat 2000-2,500 calories a day, which ,means to lose weight, the ideal calories expended daily would have an extra 500 which makes the range 2500-3000. The mean value of the sample population is 2304 and this means a good number of the sample size do not expend enough calories to lose up to 1LB weekly.
* The average amount of sedentary minutes is 991.2, this is quite high and close to the maximum (1440 minutes). Thus, it can be inferred that this sample population is more sedentary than active.


```r
sleep_day %>%  
  select(TotalSleepRecords,
  TotalMinutesAsleep, 
  TotalMinutesAwake,
  TotalTimeInBed) %>%
  summary()
```

The following can be inferred from this summary:

* According to the CDC, the healthy amount of time for sleep for an Adult is 7-9 hours (420-630 minutes). The median value is 13 minutes more than the minimum amount of time necessary for good sleep. More than half of the sample population get decent sleep.
* There is a wide gap between the 3rd quadrant and maximum values of the total minutes awake indicating 1 or more outliers in the sample population.


### 3.7 Adding Category Columns for Visualization


#### 3.7.1 Adding New Columns: Using 'weekdays()'

Both dataframes have date columns but for the purpose of visualization, new columns showing the days of the week the dates fall on would be added.


```r
daily_activity$weekday <- weekdays(daily_activity$ActivityDate)

sleep_day$weekday <- weekdays(sleep_day$SleepDay)
```

#### 3.7.2 Categorizing Existing Columns into New: Using 'mutate()'

* 'daily_activity' dataframe: Steps

Using the 10,000-step goal as an average between those that may simply want to maintain fitness level and those actively trying to lose weight, the 'TotalSteps' column is categorized into 3 sections.


```r
daily_activity <- daily_activity %>%
  mutate(StepsCategory = case_when(
    TotalSteps < 7500 ~ "Sedentary",
    TotalSteps <= 15000 ~ "Active",
    TotalSteps > 15000 ~ "Very Active"))
```

* 'daily_activity' dataframe: Calories

The average American adult consumes 2000-2500 calories per day, to lose 1LB, they would have to expend an extra 500, the calories expended during daily activity is categorised into 3 sections: those whose logged figures would amount to losses of less than 1LB, those that would lose 1-2LB weekly (the recommended amount for healthy weight loss), and those that would lose more than 2LB weekly.


```r
daily_activity <- daily_activity %>%
  mutate(CaloriesExpended = case_when(
    Calories < 2500 ~ ">1LB loss",
    Calories <= 3500 ~ "1-2LB loss",
    Calories > 3500 ~ "<2LB loss"))
```

* 'sleep_day' dataframe

The CDC recommends 7-9 hours of sleep per night for adults. This amounts to 420-540 minutes. Using these figures as constraints, the 'sleep_day' column is categorised into 3 sections.

```r
sleep_day <- sleep_day %>%
  mutate(QualityOfSleep = case_when(
    TotalMinutesAsleep < 420 ~ "Undersleeping",
    TotalMinutesAsleep <= 540 ~ "Good Quality Sleep",
    TotalMinutesAsleep > 540 ~ "Oversleeping"))
```
